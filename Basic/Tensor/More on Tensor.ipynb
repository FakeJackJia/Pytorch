# More on tensor

```python
import torch
```

```python
# filling
print(torch.full((2, 3), 10)) # return a tensor with specified size with all values the same as specified
```

```python
a = torch.rand(2, 3)
b = torch.rand(2, 3)

print(a)
print(b)
```

```python
# comparison
print(torch.eq(a, b)) # return a tensor containing the result when the corresponding elements are equal or not
print(torch.equal(a, b)) # return true/false when size and elements are equal

print(torch.ge(a, b)) # greater or equal matrix results
print(torch.gt(a, b)) # greater matrix results
print(torch.le(a, b)) # less or equal matrix results
print(torch.lt(a, b)) # less matrix results
```

```python
# sort
a = torch.tensor([[1, 4, 4, 3, 5], [2, 3, 1, 3, 5]])
print(a.shape)
print(torch.sort(a, dim=0, descending=True)) # sort based on first dimension (2 in this case)
print(torch.sort(a, dim=1, descending=True)) # sort based on second dimension (5 in this case)
```

```python
# topk
a = torch.tensor([[1, 4, 4, 3, 5], [2, 3, 1, 3, 5]])
print(a.shape)
print(torch.topk(a, k=2, dim=1)) # choose the top k element based on first dimension
print(torch.kthvalue(a, k=2, dim=1)) # choose the kth smallest element based on first dimension
```

```python
a = torch.rand(2, 3)
print(a)
print(torch.isfinite(a)) # whether element is finite or not matrix results
print(torch.isfinite(a / 0))
print(torch.isinf(a)) # whether element is infinite or not matrix results
print(torch.isinf(a / 0))

import numpy as np
a = torch.tensor([1, 2, np.nan])
print(torch.isnan(a))
```

```python
# Trigonometric functions
# others are similarly implemented
a = torch.ones(2, 3)
b = torch.cos(a) # use radian
print(b)
```

```python
# statistic
a = torch.rand(2, 2)
print(a)
print(torch.mean(a, dim=0))
print(torch.sum(a, dim=0))
print(torch.prod(a, dim=0))
```

```python
print(torch.argmax(a, dim=0)) # return the index of max element
print(torch.argmin(a, dim=0)) # return the index of min element
```

```python
print(torch.std(a))
print(torch.var(a))
# return a tuple that contains the mode value of each row
print(torch.mode(a)) # by default, dim is the last dimension of input tensor
print(torch.median(a)) # when there are even number of elements, the lower one is returned
```

```python
a = torch.rand(2, 2) * 10
print(a)
print(torch.histc(a, 6, 0, 0)) # 6 is the number of bin, and first 0 is the min and second 0 is the max
# this will divides the data according to the min and max into 6 intervals
# and calculate the frequency of elements occured in each interval
```

```python
a = torch.randint(0, 10, [10])
print(a)
# bincount only works for one dimensional tensor
print(torch.bincount(a)) # count the frequency of certain values from min up to max
```

```python
# random sampling
torch.manual_seed(1) # seed allows the result of random sampling the same
mean = torch.rand(1, 2)
std = torch.rand(1, 2)
print(torch.normal(mean, std))
```

```python
# norm
a = torch.rand(2, 1)
b = torch.rand(2, 1)

print(a, b)
# p indicates p-norm
print(torch.dist(a, b, p=1))
print(torch.dist(a, b, p=2))
print(torch.dist(a, b, p=3))

print(torch.norm(a))
print(torch.norm(a, p='nuc')) # nuclear norm: the sum of singular values of matrix
```

**EVD and SVD**
*   **EVD** requires matrix to be a **square matrix that can be diagonalized**
*   Otherwise use **SVD**

**Note**: **EVD** is usually used in **PCA** and **SVD** is usually used in **LDA**

```python
a = torch.rand(2, 2)
print(a)
print(torch.svd(a))
```

```python
# clamp
a = torch.rand(2, 2) * 10

print(a)
print(a.clamp(1, 5)) # clamps all elements within the range of [min, max]
# if < min, it becomes min
# if in range, it stay the same
# if > max, it becomes max
```

## Select index

```python
# where
a = torch.rand(4, 4)
b = torch.rand(4, 4)
print(a, b)
print(torch.where(a > 0.5, a, b)) # this indicates if condition meets, select value from a, if not select from b
print(torch.where(a > 0.5, 1, 0))
```

```python
# index_select
a = torch.rand(4, 4)
print(a)
print(torch.index_select(a, dim=0, index=torch.tensor([0, 3, 2]))) # this forms new tensor by indexing the tensor along dimension dim using entries in index
# Note: the dimth dimension of new tensor will be the length of index while others the same
```

```python
# gather
a = torch.linspace(1, 16, 16).view(4, 4)
print(a)
# Note: the result matrix is the size of index and index size need to be less than or equal to the original size
# Note; each value is retrieved from the each index on that dim
print(torch.gather(a, dim=0, index=torch.tensor([[0, 1, 1, 1], [0, 1, 2, 3]]))) # gather values along dimension dim
```

```python
# mask
print(a)
print(torch.masked_select(a, torch.gt(a, 8))) # return a 1D tensor according to the condition mask where mask is the BoolTensor
```

```python
# take
print(a)
print(torch.take(a, index=torch.tensor([0, 15, 13, 10]))) # take will treat the original tensor as 1D tensor and return 1D tensor with values of specified index
```

```python
# nonzero
a = torch.tensor([[0, 1, 0, 4, 0], [0, 2, 0, 0, 1]])
print(torch.nonzero(a)) # return the index of nonzero element in tensor
```

## Concatenation

```python
# cat
a = torch.zeros(2, 4)
b = torch.ones(2, 4)

print(a)
print(b)
print(torch.cat((b, a), dim=0)) # concatenate tensors based on the dim dimension (order matters)
# Note: result tensor has the same dimension
```

```python
# stack
a = torch.linspace(1, 6, 6).view(2, 3)
b = torch.linspace(7, 12, 6).view(2, 3)

print(a)
print(b)
out = torch.stack((a, b), dim=0) # treat each tensor as individual and create a new tensor along the new dim
print(out)
print(out.shape)
print(out[0, :, :]) # output a
print(out[1, :, :]) # output b

out = torch.stack((a, b), dim=2)
print(out)
print(out.shape)
print(out[:, :, 0]) # output a
print(out[:, :, 1]) # output b
```

## Split

```python
# chunk
a = torch.rand(3, 4)
print(a)
print(torch.chunk(a, 2, dim=0)) # split the tensor into 2 parts by averaging on dimension 0
# Note: if not evenly splitted, the last chunk will have fewer
```

```python
# split
print(a)
print(torch.split(a, 1, dim=0)) # split the tensor into several tensors while each tensor has size 1 of dimension dim
# Note: similarly, the last one will have fewer if not evenly
```

## Transformation

```python
a = torch.rand(2, 3)
print(a)
print(torch.reshape(a, (3, 2))) # reshape
print(torch.t(a)) # transpose
print(torch.transpose(a, 0, 1)) # tranpose between dimension 0 and 1
```

```python
a = torch.rand(1, 2, 3)
print(a)
print(torch.squeeze(a)) # remove dimension that has size 1
print(torch.unsqueeze(a, -1).shape) # add a new dimension of size 1 at last position
```

```python
print(torch.unbind(a, dim=2)) # remove specified dim

print(a)
print(torch.flip(a, dims=[1, 2])) # dims indicates collection of dim to flip
```

```python
a = torch.rand(2, 3)
print(a)

out = torch.rot90(a, 1) # rotate 90 degree k times clockwise if k is negative otherwise unclockwise
print(out)
print(out.shape)
```
