# More on tensor

```python
import torch
```

```python
a = torch.rand(2, 3)
b = torch.rand(2, 3)

print(a)
print(b)
```

```python
# comparison
print(torch.eq(a, b)) # return a tensor containing the result when the corresponding elements are equal or not
print(torch.equal(a, b)) # return true/false when size and elements are equal

print(torch.ge(a, b)) # greater or equal matrix results
print(torch.gt(a, b)) # greater matrix results
print(torch.le(a, b)) # less or equal matrix results
print(torch.lt(a, b)) # less matrix results
```

```python
# sort
a = torch.tensor([[1, 4, 4, 3, 5], [2, 3, 1, 3, 5]])
print(a.shape)
print(torch.sort(a, dim=0, descending=True)) # sort based on first dimension (2 in this case)
print(torch.sort(a, dim=1, descending=True)) # sort based on second dimension (5 in this case)
```

```python
# topk
a = torch.tensor([[1, 4, 4, 3, 5], [2, 3, 1, 3, 5]])
print(a.shape)
print(torch.topk(a, k=2, dim=1)) # choose the top k element based on first dimension
print(torch.kthvalue(a, k=2, dim=1)) # choose the kth smallest element based on first dimension
```

```python
a = torch.rand(2, 3)
print(a)
print(torch.isfinite(a)) # whether element is finite or not matrix results
print(torch.isfinite(a / 0))
print(torch.isinf(a)) # whether element is infinite or not matrix results
print(torch.isinf(a / 0))

import numpy as np
a = torch.tensor([1, 2, np.nan])
print(torch.isnan(a))
```
