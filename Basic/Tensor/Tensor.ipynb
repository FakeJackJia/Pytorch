# Tensor

**Tensor** is mathematical term that **generalizes scalars, vectors and matrices to higher dimensions**.

image.png

## Basic Tensor

```python
import torch
```

```python
# directly initialize tensor by specifying the values
a = torch.tensor([[1, 2], [3, 4]])
print(a)
print(a.type())
```

```python
# Can also only specify the shape of tensor
a = torch.Tensor(2, 3) # note: values in this tensor are randomly initialized
print(a)
print(a.type())
```

```python
# Tensor with all ones of shape 2x2
a = torch.ones(2, 2)
print(a)
print(a.type())
```

```python
# Tensor with ones on the diagonal and zeros otherwise of shape 2x2
a = torch.eye(2, 2)
print(a)
print(a.type())
```

```python
# Tensor with all zeros of the same shape as a
b = torch.zeros_like(a)
print(b)
print(b.type())
```

## Random Tensor

```python
# Tensor with random values (0 to 1) of shape 2x2
a = torch.rand(2, 2)
print(a)
print(a.type())
```

```python
# Tensor with each value obtained from their corresponding normal distribution of mean=0 and random std obtained
a = torch.normal(mean=0.0, std=torch.rand(5))
print(a)
print(a.type())
```

```python
# Tensor with each value obtained from uniform distribution of shape 2x2
a = torch.Tensor(2, 2).uniform_(-1, 1)
print(a)
print(a.type())
```

## Sequence Tensor

```python
# Tensor from 0 to 9 with step 1
a = torch.arange(0, 10, 1)
print(a)
print(a.type())
```

```python
# Tensor from 2 to 10 with 3 numbers with the same interval
a = torch.linspace(2, 10, 3)
print(a)
print(a.type())
```

```python
# Tensor from sequence from 0 to 9 and then shuffle
a = torch.randperm(10)
print(a)
print(a.type())
```

## Attribute

Every tensor has **three attributes**: **torch.dtype, torch.device, and torch.layout**
*   **torch.device** indicates the **device** where the torch.Tensor Object located after it is created
*   **torch.layout** indicates memory layout of tensor
  *   In detail, it shows **how many nonzero element is in the tensor**, if there are only few, then it is called **Sparse tensor**; otherwise, it is a **Dense tensor**





```python
dev = torch.device("cpu")
a = torch.tensor([2, 2], dtype=torch.float32, device=dev)
print(a)
```

```python
# i indicates the position of nonzero element
i = torch.tensor([[0, 1, 2], [0, 1, 2]]) # pos (0, 0), (1, 1), (2, 2)
# v indicates the value on it corresponding position
v = torch.tensor([1, 2, 3]) # e.g 1 on (0, 0)
# create a sparse tensor
a = torch.sparse_coo_tensor(i, v, (4, 4), dtype=torch.float32, device=dev) #(4, 4) indicates the size of tensor
print(a)
```

```python
# convert a dense tensor to a sparse tensor
a.to_dense()
```

## Arithmetic operation

```python
# addition
a = torch.rand(2, 3)
b = torch.rand(2, 3)

# the below three is the same
print(a + b)
print(a.add(b))
print(torch.add(a, b))

print(a)
print(a.add_(b)) #add_ will also modify a
print(a)
```

```python
# subtraction
# the below three is the same
print(a - b)
print(a.sub(b))
print(torch.sub(a, b))

print(a.sub_(b)) #similiarly, sub_ will also modify a
print(a)
```

```python
# multiplication
# Note: this is element wise! It is not matrix multiplication
print(a * b)
print(torch.mul(a, b))
print(a.mul(b))

print(a.mul_(b)) # modify a
print(a)
```

```python
# division
print(a / b)
print(torch.div(a, b))
print(a.div(b))

print(a.div_(b))
```

```python
# matrix multiplcation
a = torch.ones(2, 1)
b = torch.ones(1, 2)

print(a @ b)
print(a.matmul(b))
print(torch.matmul(a, b))
print(torch.mm(a, b))
print(a.mm(b))
```

```python
# high dimensional tensor
# Note: matrix multiplication will only applied to the lowerest 2 dimension
# the dimension beyond that need to have the same dimension
a = torch.ones(1, 2, 3, 4) # Note: the lower 2 dimension still need to satisfy the rule of matrix multiplication
b = torch.ones(1, 2, 4, 3)
print(a.matmul(b))
print(a.matmul(b).shape)
```

```python
# power
a = torch.tensor([1, 2])
print(a.pow(3))
print(a**3)

print(a.pow_(3)) # modify a
```

```python
# exp
a = torch.tensor([1, 2], dtype=torch.float32)
print(torch.exp(a))
print(torch.exp_(a)) # modify a

print(a.exp())
print(a.exp_()) # modify a
```

```python
# log
a = torch.tensor([10, 2], dtype=torch.float32)
print(torch.log(a))
print(torch.log_(a)) # modify a

print(a.log())
print(a.log_()) # modify a
```

```python
# sqrt
a = torch.tensor([10, 2], dtype=torch.float32)
print(torch.sqrt(a))
print(torch.sqrt_(a)) # modify a

print(a.sqrt())
print(a.sqrt_()) # modify a
```

```python
# division
a = torch.rand(2, 2)
a = a * 10
print(a)
print(torch.floor(a)) # floor divsion
print(torch.ceil(a)) # ceil division
print(torch.round(a)) # rounding division
print(torch.trunc(a)) # keep integer part
print(torch.frac(a)) # keep fraction part
print(a % 2) # mod 2
```

**Broadcast mechanism**: when two tensors **have different dimensions** and you still want to **perform arithmetic operation on them**, the one will with the smaller size will be **expanded to the same dimension as the other one**.
Two criterias need to be satisfied:
*   Every tensor has **at least one dimension.**
*   When iterating over the dimension sizes, **starting at the trailing dimension**, the dimension sizes must either be **equal, one of them is 1, or one of them does not exist**.

```python
a = torch.rand(2, 1, 1, 3)
b = torch.rand(4, 2, 3)
c = a + b # result size on each dimension is the largest size from each dimension
print(a)
print(b)
print(c)
print(c.shape)
```
